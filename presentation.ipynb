# Regression

```python
df_not_smoker = df[df["smoker"] == "No"]
df_smoker = df[df["smoker"] == "Yes"]

model1 = smf.ols("tip ~ total_bill" , data = df_not_smoker).fit()
model2 = smf.ols("tip ~ total_bill" , data = df_smoker).fit()
model3 = smf.ols("tip ~ total_bill + smoker + total_bill:smoker", data = df).fit() #mit dummyvariable
```

```python
Stargazer([model1, model2, model3])
```
